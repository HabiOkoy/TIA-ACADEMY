```python
def printme(str):
 "This prints a passed string into this function"
 print (str)
 return;
```

```python
mylist = [10,20,30];
changeme( mylist );
print ("Values outside the function: ", mylist)
```

```python
def printinfo( name, age ):
 "This prints a passed info into this function"
 print ("Name: ", name)
 print ("Age: ", age)
 return;

printinfo( age=23, name="okoy" )
```

```python
def printinfo( arg1, *vartuple ):
 "This prints a variable passed arguments"
 print ("Output is: ")
 print (arg1)
 for var in vartuple:
     print (var)
 return;
```

```python
printinfo( 10 )
printinfo( 70, 60, 50 )
```
