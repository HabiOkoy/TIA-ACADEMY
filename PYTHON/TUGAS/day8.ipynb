### Operator Arimatik

```python
a = 5
b = 10

# Addition
print('a + b = ', a + b)

# Subraction
print('a - b = ', a - b)

# Multiplication
print('a * b = ', a * b)

# Division
print('a / b = ', a / b)

# Modulus
print('a % b = ', a % b)

# Exponent
print('a ** b = ', a ** b)

# Floor Division
print('a // b = ', a // b)
```

### while loop

```python
i = 1
while i < 15:
    print(i)
    if i == 9: break
    i += 1
```

```python
# nested looping
films = ["harrypoter", "spiderman", "batman", "john wick"]
rating = ["Bagus", "Biasa", "Jelek"]
for i in rating:
    for j in films:
        print(i, j)
```
