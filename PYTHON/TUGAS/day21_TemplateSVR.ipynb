```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

## Import dataset

```python
dataset = pd.read_csv('/content/gdrive/MyDrive/Python TIA/ml-1m/DataListrik.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X)
```

```python
print(y)
```

```python
y_train=y_train.reshape(len(y_train),1)
print(y)
```

## Feature scalling standardscale

```python
sc_X = StandardScaler()
sc_y = StandardScaler()
X_train = sc_X.fit_transform(X_train)
y_train= sc_y.fit_transform(y_train)
```

```python
print(X_train)
```

```python
print(y_train)
```

## Training the SVR model on the Training set

```python
regressor = SVR(kernel='rbf')
#X,y disini adalah X,y yang sudah mengalami transformasi scalling
regressor.fit(X_train,y_train)
```

```python
print(y_test)
```

```python
y_test_scale = sc_y.transform(y_test.reshape(-1,1))
y_pred_scale = sc_y.inverse_transform(regressor.predict(sc_X.transform(X_test)).reshape(-1,1))
y_test_scale = y_test.reshape(len(y_test),1)
#print(sc_y.inverse_transform(regressor.predict(sc_X.transform([[6.5]])).reshape(-1,1))) #hasilnya menjadi yg adad di feature scalling


```

## Predicting the Test set results

```python
print(np.concatenate((y_pred_scale,y_test_scale),1))
```
