```python
#highly rated movies
highly_rated = movie_ratings_data.groupby('title').agg({'rating' : [np.size, np.mean]})
highly_rated
```

```python
#highly_rated = movie_ratings_data.groupby('title').agg({'rating' : [np.size, np.mean]})
highly_rated_sorted = highly_rated.sort_values(('rating','size'), ascending=False)
highly_rated_sorted
```

```python
# Load Data Users
dnames = ['userid', 'gender', 'age', 'occupation', 'zip-code']
users_data = pd.read_table('/content/gdrive/MyDrive/Python TIA/ml-1m/users.dat', sep='::', header=None, names=dnames, engine='python', encoding='latin=1')
print('movie shape: ',movies_data.shape)
movies_data.info()
```

```python
users_data
```

```python
#menampilkan age berdasarkan range 
#kurang tepat
age_counts = users_data['age'].value_counts()
age_counts
```

```python
age_dict = {1: '0-18',
            18: '18-24',
            25: '25-34',
            35: '35-44',
            45: '45-49',
            50: '50-55',
            56: '56+'}

# Replace nilai age dengan label
users_data['age'] = users_data['age'].replace(age_dict)

# Tampilkan jumlah setiap label pada kolom age
users_data.head(5)
```

```python
# 3 grafik histogram
#NO 3
age_counts = users_data['age'].value_counts()
age_counts
```

```python
age_counts_sorted = age_counts.sort_index()
age_counts_sorted
```

```python
import matplotlib.pyplot as plt
plt.bar(age_counts_sorted.index, age_counts_sorted.values)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

PIVOT TABLE

```python
data_sales = {
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'A', 'B', 'B'],
    'Sales': [100, 200, 150, 250, 120, 180, 130, 220]
}

df_sales = pd.DataFrame(data_sales)
df_sales
```

```python
#membuat tabel baru
#kolom: product A, B
#baris: Region East, West, South, West

df_sales_pivot = df_sales.pivot_table(columns='Product', index='Region', aggfunc='sum').fillna(0)
df_sales_pivot
```
