```python
# dependent    : var salary     ==> y (vektor)
# independent  : var experience ==> x (matriks n x n)
# baris,kolom
# x = dataset.iloc[0:29,0] # cara 1    # semua baris , semua colomn kecuali salary
x = dataset.iloc[:,:-1]    # cara 2    # mengambil semua baris,kolom kecuali yang terakhir yaitu salary
y = dataset.iloc[:,-1]     # mengambil semua baris,kolom  yang terakhir saja (y)
```
