```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x,y)
plt.xlabel('Pulse')
plt.ylabel('Calories burn')
plt.show()
```

plt.plot() -> isi parameter dari plot() adalah variabel x dan y. parameter ketiga adalah buat edit bentuk plot, contoh : plt.plot(x,y,marker='o') atau bisa juga melakukan default yaitu plt.plot(x,y) maka akan menghasilkan line

## Menamakan line x dan y

```python
plt.plot(x,y)
plt.title('Data orang leveling')
plt.xlabel('Jam')
plt.ylabel('Level character')
plt.show()
```

```python
plt.plot(x,y)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'red','size':15}
plt.title('Data orang leveling',fontdict = font1)
plt.xlabel('Jam',fontdict = font2)
plt.ylabel('Level character',fontdict =font2)
plt.show()
```

## Display Multiple plot

```python
#plot 1
x1 = np.array([2,1,5,8])
y2 = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x1,y2)

#plot 2
x = np.array([3,0,6,9])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

plt.subplot() -> subplot() memiliki 3 parameter yaitu nrows,ncols,index (1,2,1) -> artinya adalah memiliki 1 rows, 2 coloumn jadi nanti akan dibagi 1/2 dan index 1

```python
plt.subplot(2,1,1)
plt.plot(x1,y1)

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

## Pie

```python
y = np.array([35,25,15,10])
plt.pie(y)
plt.show()
```

```python
y = np.array([35,25,15,10])
mylabel =['Axe','Sword','Katana','Bow']

plt.pie(y,labels=mylabel,startangle=90)
plt.show()
```

```python
y = np.array([35,25,15,10])
mylabel =['Axe','Sword','Katana','Bow']
myexplode = [0.2,0,0,0]
mycolor = ['black','pink','navy','blue']

plt.pie(y,labels=mylabel,startangle=90,colors=mycolor,explode=myexplode)
plt.legend(title = "Senjata favorit :")
plt.show()
```
