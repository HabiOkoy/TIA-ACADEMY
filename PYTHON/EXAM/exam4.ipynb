## Importing the dataset

```python
#memisahkan antara variabel Independen da variabel dependen

X = dataset.iloc[:,:-1] #semua kolom kecuali PE
y = dataset.iloc[:,-1] # hanya PE saja
```

```python
print(X)
```

```python
print(y)
```

## Spliting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)  # data pengujian 20%
```

## Training the Polynomial regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree=4)
X_poly_train = poly_reg.fit_transform(X_train)
X_poly_test = poly_reg.transform(X_test)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_train, y_train)
```

## Predicting the test set result

```python
# predict data based on X tertentu 
y_pred = lin_reg_2.predict(X_poly_test)
y_pred
```

```python
# performance  metrica
# bandingkan antara y_pred dan y_test
# y_pred_s sebagai kolom pertama, y_test_s sebagai kolom ke dua
y_pred_s = y_pred.reshape(len(y_pred),1)
y_test_s = np.array(y_test).reshape(len(y_test),1)
print(np.concatenate((y_pred_s, y_test_s),1))
```

```python
# performance metric
y_pred_polynom = lin_reg_2.predict(X_poly_test)
comparison_df = pd.DataFrame({'y_pred_s': y_pred_polynom, 'y_test': y_test})
print(comparison_df)
```
