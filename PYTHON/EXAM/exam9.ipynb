```python
import pandas as pd
from mlxtend.frequent_patterns import association_rules, apriori
```

```python
df = pd.read_csv('/content/gdrive/MyDrive/Python TIA/ml-1m/TV Shows - Association Rule Learning - TV Shows - Association Rule Learning.csv')
```

```python
df = df.drop(['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8',
              'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13',
              'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
              'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23',
              'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28',
              'Unnamed: 29', 'Unnamed: 30', 'Unnamed: 31'], axis=1)
```

```python
df_encoded = pd.get_dummies(df)
```

```python
frequent_itemsets = apriori(df_encoded, min_support=0.003, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2)
rules = rules[rules['lift'] >= 3]
```
